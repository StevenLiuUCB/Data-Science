```python
import numpy as np
import pandas as pd
import re
```

```python
data = pd.read_csv("data_new.csv")
```

```python
data = data.drop(['unknown1','unknown2','unknown3','unknown4'],1)
```

```python
for i in range(len(data)):
    if len(data.elevator[i].split()[0])>1:
        data.iloc[i, 12] = '无'
    else :
        data.iloc[i, 12] = data.elevator[i].split()[0]
```

```python
vroom = []
bedroom = []
bathroom = []
for i in range(len(data)):
    k = re.findall(r'[0-9]',data.room[i])
    try:
        vroom.append(k[0])
        bedroom.append(k[1])
        bathroom.append(k[2])
    except:
        data = data.drop(i)
```

```python
data['vroom'] = vroom
data['bedroom'] = bedroom
data['bathroom'] = bathroom
```

```python
data = data.drop('room', axis=1)
```

```python
data = data.reset_index(drop=True)
```

```python
area = []
c = 0
for i in range(len(data)):
    k = re.findall(r'[0-9]+',data.area[i])
    area.append(k[0])
    c+=1
c
```

```python
data['area'] = area
```

```python
price = []
for i in range(len(data)):
    try:
        k = re.findall(r'[0-9]+',data.price[i])
        price.append(k[0])
    except:
        data = data.drop(i)
```

```python
data['price'] = price
```

```python
c=[]
for i in range(len(data)):
    a = re.findall(r'[0-9]+', data.time[i])[0]
    if len(a)>4:
        data = data.drop(i)
    else:
        c.append(2018-int(a))
```

```python
data = data.reset_index(drop=True)
```

```python
data['time'] = c
```

```python
data.drop('web', 1, inplace=True)
```
